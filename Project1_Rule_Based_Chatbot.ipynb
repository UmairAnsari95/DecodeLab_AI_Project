## Goal
Create a simple rule-based chatbot that responds to predefined user inputs.

## Key Requirements
- Handle greetings and exit commands

## Key Skills
Control flow, decision-making logic, basic AI concepts

## Step 1: Input & Sanitization

Before matching user input against our knowledge base, we clean it up:
- Convert to lowercase
- Strip leading/trailing whitespace

This ensures `"Hello"`, `"hello "`, and `"HELLO"` are all treated the same way.

```python
def sanitize_input(raw_input):
    """Clean the raw user input: lowercase + strip whitespace."""
    return raw_input.lower().strip()
```

## Step 2: Knowledge Base (Dictionary)

Instead of a long if-elif ladder (which is slow and hard to maintain),
we use a **dictionary** for instant O(1) lookups.

The knowledge base below contains 5+ intents.

```python
responses = {
    "hello": "Hi there! How can I help you today?",
    "hi": "Hello! Nice to see you.",
    "how are you": "I'm just a program, but I'm doing great! Thanks for asking.",
    "what is your name": "I am a Rule-Based AI Chatbot built for DecodeLabs Project 1.",
    "what can you do": "I can respond to a few predefined greetings and questions.",
    "thank you": "You're welcome!",
    "thanks": "You're welcome!",
    "bye": "Goodbye! Have a great day."
}

FALLBACK_RESPONSE = "I do not understand. Could you please rephrase that?"
EXIT_COMMANDS = {"exit", "quit", "bye"}
```

## Step 3: Response Logic (The `.get()` Method)

The dictionary's `.get()` method performs the lookup and fallback
in a single atomic operation: if the key exists, return its value;
otherwise, return the fallback response.

```python
def get_response(clean_input):
    """Look up the response for the cleaned input, or return the fallback."""
    return responses.get(clean_input, FALLBACK_RESPONSE)
```

## Step 4: The Heartbeat — Continuous Loop

The chatbot runs in an infinite `while True` loop.
It keeps taking input until the user types an exit command
(`exit`, `quit`, or `bye`), at which point it breaks out cleanly.

```python
def run_chatbot():
    """Main chatbot loop."""
    print("Bot: Hello! I'm a rule-based chatbot. Type 'exit' or 'quit' to end our chat.")

    while True:
        raw_input_text = input("You: ")
        clean_input = sanitize_input(raw_input_text)

        if clean_input in EXIT_COMMANDS:
            print("Bot:", responses.get("bye", "Goodbye!"))
            break

        reply = get_response(clean_input)
        print("Bot:", reply)
```

## Step 5: Run the Chatbot

Run the cell below to start chatting. Type `exit`, `quit`, or `bye` to stop.

```python
run_chatbot()
```
